```python
import matplotlib.pyplot as plt
import numpy as np
import csv, os, statistics
from typing import List
import utils as ut

##### CONFIG HERE #####
RESULT_FOLDER="alpha"
CONCUR_USER = 1
PERCENTAGE = 99
#######################
RESULT_FOLDER="result/"+RESULT_FOLDER



barWidth = 0.3

samesideMean = []
diffsideMean = []
samesideStdev = []
diffsideStdev = []

rt = ut.readFromCSVFile(f"{RESULT_FOLDER}/VanillaEdgeEdge_{CONCUR_USER}.csv", PERCENTAGE)
samesideMean.append(statistics.mean(rt))
samesideStdev.append(statistics.stdev(rt))

rt = ut.readFromCSVFile(f"{RESULT_FOLDER}/VanillaEdgeCloud_{CONCUR_USER}.csv", PERCENTAGE)
diffsideMean.append(statistics.mean(rt))
diffsideStdev.append(statistics.stdev(rt))

rt = ut.readFromCSVFile(f"{RESULT_FOLDER}/VanillaCloudCloud_{CONCUR_USER}.csv", PERCENTAGE)
samesideMean.append(statistics.mean(rt))
samesideStdev.append(statistics.stdev(rt))

rt = ut.readFromCSVFile(f"{RESULT_FOLDER}/VanillaCloudEdge_{CONCUR_USER}.csv", PERCENTAGE)
diffsideMean.append(statistics.mean(rt))
diffsideStdev.append(statistics.stdev(rt))

r1 = np.arange(len(samesideMean))
r2 = [x + barWidth for x in r1]

plt.title("Response Time according to the Relative Position\nbetween Traffic Source and KSCs + Function")
plt.bar(r1, samesideMean, width = barWidth, color = 'blue', edgecolor = 'black', yerr=samesideStdev, capsize=5, label='KSCs on the same side')
plt.bar(r2, diffsideMean, width = barWidth, color = 'cyan', edgecolor = 'black', yerr=diffsideStdev, capsize=5, label='KSCs on the diff side')
plt.xticks([r + barWidth/2 for r in range(len(samesideMean))], ['Traffic from Edge', 'Traffic from Cloud'])
plt.ylabel('Location of traffic')
plt.ylabel('Response Time (ms)')
plt.legend(loc=10, bbox_to_anchor=(1.25, 0.2))

plt.show()
```
